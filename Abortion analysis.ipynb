# An analysis on abortion in India

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
file_path = './Abortion.csv'
ab = pd.read_csv(file_path)
```

```python
columns = ab.columns.to_list()
gender = columns[8]
consent = columns[24]
safety = columns[-5]
```

```python
ab = ab.replace({gender:{'Femalen':'Female','Female ':'Female','F':"Female",'FEMALE':'Female',
                'M':'Male','Male ':'Male'},
                 consent:{'Consent of the family/husband should be required\xa0':'Required',
                          'Consent of the family/husband is not required\xa0':'Not required',
                          'No comments\xa0':'No comment'}})
```

```python
ab
```

```python
sns.countplot(x=ab.loc[:,consent],hue=ab.loc[:,gender])
plt.show()
```

```python
sns.swarmplot(x=ab.loc[:,consent],y=ab.loc[:,safety],hue=ab.loc[:,gender],dodge=True,size=5)
plt.show()
```
